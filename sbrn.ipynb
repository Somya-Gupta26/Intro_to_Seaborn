<h2><b>SEABORN</b></h2>
It is built on top of Matplotlib and integrates closely with Pandas data structures, making it easier to visualize datasets stored in Pandas DataFrames.

```python
import seaborn as sns
sns.__version__
```

```python
#INTERNSHALA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#SAMPLE DATA
data={'category':["Product A","Product B","Product C","Product D"],'value':[150,200,120,250]}
df=pd.DataFrame(data)
#CREATING A BAR CHART
sns.barplot(x='category',y='value',data=df)
#CUSTOMZING THE PLOT
sns.set_style('whitegrid')     #set background style
# sns.set_color_codes('pastel')       #change bar colors to pastel palette
#ADDING LABELS AND TITLE
plt.xlabel("Product Category")
plt.ylabel("Purchase Values")
plt.title("Customer Purchase by Product Category")
plt.show()
```

```python
#7th AUGUST
df=sns.load_dataset('tips')     #Seaborn has some in-built datsets in it's Github repository that are being downloaded and loaded just using a load_dataset function.
df
```

```python
sns.scatterplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
df['day'].value_counts()
```

```python
sns.countplot(df['day'])
plt.show()
```

```python
sns.countplot(df['total_bill'])
plt.show()
```

<h4><b>NUMERICAL PLOT </b></h4>(both col are numerical)

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='time',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='total_bill',y='size',data=df)
plt.show()
```

```python
df[['total_bill','tip','size']].corr()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',kind='line')       #kind accepts string that describes what plot we want to draw (only allows line or scatter)      #by default it is "scatter"
plt.show()
```

<h4><b>CATEGORICAL PLOT </b></h4>(one col is numerical and other one is categorical)

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex')      
plt.show()
#Below chart is known as STRIP CHART
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='violin')      #kind allows 'strip', 'swarm', 'box', 'violin', 'boxen', 'point', 'bar', 'count' types of plot      #by default it is "STRIP"  
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='bar') 
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='swarm') 
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='box') 
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='boxen') 
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,kind='point') 
plt.show()
```

```python
sns.catplot(x='smoker',data=df,kind='count') 
plt.show()
```

<h4><b>DIS PLOT</b></h4>

```python
sns.displot(df['total_bill'],ec='k',kde=True,color='g')       #it's like a histogram
plt.show()
```

```python
sns.histplot(df['total_bill'],ec='black')
plt.show()
```

<h4><b>KERNEL DENSITY ESTIMATE(KDE) PLOT</b></h4>

```python
sns.kdeplot(df['total_bill'])       
plt.show()
```
